```python
# import dependencies

import pandas as pd
from config import api_key
import requests
import os
import math
from datetime import datetime
from config import finn
import finnhub
finnhub_client = finnhub.Client(api_key=finn)
import time
import datetime
now = datetime.datetime.now()
import yfinance as yf
import numpy as np
from scipy.stats import norm
import yfinance as yf
```

```python
# Import stock symbols

symbols = pd.read_csv("Stocks.csv")["Tickers"].tolist()
```

```python
sigma = {}


for ticker in symbols:
    quote = f"https://api.tdameritrade.com/v1/marketdata/{ticker}/quotes"

    #uses api key to extract data
    params = {'apikey':api_key}


    content = requests.get(url = quote, params=params)

    data = content.json()
    
    if len(data) == 0:
        div = 0
    else:
        div = data[ticker]['divYield']

    
    sigma[ticker] = {'Sigma': 0, 'Mu': 0}


    # Set the start and end date
    start_date = '1990-01-01'
    end_date = '2023-01-01'

    # Get the data
    data = yf.download(ticker, start_date, end_date)
    
    #Convert to dataframe
    data_df = pd.DataFrame(data)
    
    # Shift close by one day

    data_df["Previous_Close"] = data_df["Close"].shift(1)

    # Calculate Rn

    data_df["R"] = np.log((data["Close"] - data["Previous_Close"]) / data["Previous_Close"] + 1)
    
    data_df = data_df.dropna()

    # N days
    N = len(data_df)

    # Daily mean

    u = data_df["R"].sum() / N

    # Difference between R and u

    data_df["R - u"] = (data_df["R"] - u) ** 2

    # Calculate Volatility

    o = math.sqrt(data_df["R - u"].sum() / (N - 1))
    
    sigma[ticker]['Sigma'] = o * math.sqrt(252)
    
    sigma[ticker]['Mu'] = u
    
    sigma[ticker]['Div'] = div
    
    time.sleep(0.4)
    
sigma_df = pd.DataFrame.from_dict(sigma, orient="index")

sigma_df.to_csv("Sigma.csv")
```
